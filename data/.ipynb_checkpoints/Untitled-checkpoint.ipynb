```python
import pandas as pd
import numpy as np
from datetime import datetime
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/2020_coronavirus_pandemic_in_South_Africa#Cases'
url = url
request = requests.get(url)
website = request.text
        
soup = BeautifulSoup(website, features = "lxml")
table = soup.find('table', {'class': 'wikitable'})
```

```python
col_names = list()
for node in range(26):
    value = ', '.join(table.findAll('th')[node].text.split())
    col_names.append(value)
            
# col_names.insert(11, 'New cases')
# col_names.insert(13, 'New deaths')
col_names
```

```python
columns = 18
rows = len(table.findAll('tr'))

df = pd.DataFrame(index = range(rows))

for column in range(columns):
    values = list()
    for each_row in range(rows):
        row = table.findAll('tr')[each_row].findAll('td')
        if len(row) == columns:
            value = ', '.join(row[column].text.split())
            if ']' in value:
                values.append(value.split(']')[1])
            elif value == '?':
                values.append(np.nan)
            elif value and ']' not in value:
                values.append(value)
            else:
                values.append(0)

            df['{}'.format(column)] = pd.Series(values)

# df.columns = col_names
# df = df.dropna(how = 'all')
# df['Day'] = range(-1, len(df) - 1)
df
```
