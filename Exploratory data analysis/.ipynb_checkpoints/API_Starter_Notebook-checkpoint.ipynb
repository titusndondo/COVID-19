The data comes from a huge central repository created by the team at Explore AI. The database is a collection of publicly available data delivered to the community through APIs. The code below will pull the various tables into our environment.


```python
casesGlobal = 'https://9gnht4xyvf.execute-api.eu-west-1.amazonaws.com/api/get_table/CasesGlobalView'
casesLocal = 'https://9gnht4xyvf.execute-api.eu-west-1.amazonaws.com/api/get_table/CasesLocalView'
testsLocal = 'https://9gnht4xyvf.execute-api.eu-west-1.amazonaws.com/api/get_table/Testing'
counterMeasures = 'https://9gnht4xyvf.execute-api.eu-west-1.amazonaws.com/api/get_table/CounterMeasureView'

urls = {
    'casesGlobal': casesGlobal,
    'casesLocal': casesLocal,
    'testsLocal': testsLocal,
    'counterMeasures': counterMeasures
}

headers = {
    'x-api-key': "WVllUkRA01awNNgKxGg607vl5qIvuOAN3pW9HXmD"
    }

tables = {}
for table_name, url in urls.items():
    response = requests.request("GET", url, headers = headers)
    x = response.json()
    tables[table_name] = pd.DataFrame(x).set_index('date')
    tables[table_name].index = pd.to_datetime(tables[table_name].index).date
```

*TABLE DESCRIPTIONS*

```python
print('TABLE: casesLocal')
print('Rows and columns:', tables['casesLocal'].shape)
print('Countries:', tables['casesLocal']['country_id'].nunique(), '(SOUTH AFRICA)')
print('Provinces:', tables['casesLocal']['location'].nunique())
print('Records from:', tables['casesLocal'].index.min(), 'to', tables['casesLocal'].index.max())
print()
print('TABLE: testsLocal')
print('Rows and columns:', tables['testsLocal'].shape)
print('Countries:', tables['testsLocal']['country'].nunique(), '(SOUTH AFRICA)')
print('Records from:', tables['testsLocal'].index.min(), 'to', tables['testsLocal'].index.max())
print()
print('TABLE: counterMeasures')
print('Rows and columns:', tables['counterMeasures'].shape)
print('Number of measures implemented to date:', tables['counterMeasures']['measure'].nunique(), 'in', tables['counterMeasures']['country'].nunique(), 'countries around the world')
print('Records from:', tables['counterMeasures'].index.min(), 'to', tables['counterMeasures'].index.max())

print()
print('TABLE: casesGlobal')
print('Rows and columns:', tables['casesGlobal'].shape)
print('Number of countries recorded:', tables['casesGlobal']['country'].nunique())
print('Records from:', tables['casesGlobal'].index.min(), 'to', tables['casesGlobal'].index.max())
```

```python
tables['casesLocal'].to_csv('casesLocal.csv')
tables['testsLocal'].to_csv('testsLocal.csv')
tables['counterMeasures'].to_csv('counterMeasures.csv')
tables['casesGlobal'].to_csv('casesGlobal.csv')
```

```python
confirmed_cases = tables['casesGlobal'].loc[:, ['country', 'confirmed', 'confirmed_daily']]
confirmed_cases = confirmed_cases.pivot(values = 'confirmed', columns = 'country')

df_dict = {}
for country in confirmed_cases.columns:
    cases = confirmed_cases[country].dropna()
    cases.index = range(1, len(cases) + 1)
    df_dict[country] = cases
confirmed_cases_by_day = pd.DataFrame({k: pd.Series(l) for k, l in df_dict.items()})
confirmed_cases_by_day.head()
```

```python
fontsize = 14
top_countries = ['South Africa','Italy', 'United Kingdom']

confirmed_cases[top_countries].plot(figsize = (18, 5), fontsize = fontsize)
plt.legend(fontsize = 14)


confirmed_cases_by_day[top_countries].plot(figsize = (18, 5), fontsize = fontsize)
plt.legend(fontsize = 14)
```
